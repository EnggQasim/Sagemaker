```python
import wget
import requests
from bs4 import BeautifulSoup
import io
import urllib.request
url ='https://www.cbe.org.eg/en/BankingSupervision/Pages/Circulars.aspx'
downlaod_url ='https://www.cbe.org.eg'
useragent ='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36'
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
ul = soup.findAll('li',{'class':'clearfix'})

def fn_download_url(file_url):
    file_name = file_url.split('/')
    file_name = 'c:\\data\\downloads\\' + file_name[-2] + file_name[-1]
    opener = urllib.request.URLopener()
#     opener = urllib.parse.urlencode()
    opener.addheader('User-Agent',useragent)
    file_name,headers = opener.retrieve(file_url , file_name)
    
    print(file_url,file_name)

for a in ul:
    for items in a.findAll('a'):
        file_url = downlaod_url+items['href'].replace(' ','%20')
        print(file_url)
        print('Donwload start')
        fn_download_url(safe_str(file_url))
        print('Download Complete')
```

```python
def safe_str(obj):
    try: return str(obj)
    except UnicodeEncodeError:
        return obj.encode('utf-8', 'ignore').decode('cp1252')
    return ""

safe_str(u'\'')
```
